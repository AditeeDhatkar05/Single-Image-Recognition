# Load a deep learning model

```python
mobile=tf.keras.applications.mobilenet_v2.MobileNetV2()  # deep learning pre-trained model
```

# Preprocessing of the image

```python
from tensorflow.keras.preprocessing import image
img=image.load_img(file_name,target_size=(224,224))
img
```

```python
resized_image=image.img_to_array(img)
resized_image.shape
```

```python
final_img=np.expand_dims(resized_image,axis=0)
final_img.shape
```

```python
final_img=tf.keras.applications.mobilenet_v2.preprocess_input(final_img)
final_img
```

```python
final_img.shape
```

```python
prediction=mobile.predict(final_img)
prediction
```

```python
from tensorflow.keras.applications import imagenet_utils
results=imagenet_utils.decode_predictions(prediction)
print(results)
```
